## 4.2
### the equation
### $ a^{3} + b^{3} = c^{2} $
### has integer solutions such that $ c>=b>=a>=1 $

### one solution is (a, b, c) = (2, 2, 4)

```python
import pandas as pd

def doesWork(a, b, c):
    return a**3 + b**3 == c**2

def generateTable(ar, br, cr): #ranges for each
    d={"A": [],"B": [], "C": []}
    for i in range(1, ar+1):
        for j in range(i, br+1):
            for k in range(j, cr+1):
                if doesWork(i, j, k):
                    d["A"].append(i)
                    d["B"].append(j)
                    d["C"].append(k)
    return pd.DataFrame(d)
```

```python
DF1=generateTable(100, 100, 150)
display(DF1)
```

## ok so that didnt give a ton of good results
## lets try again, but folloing form (a, b, c)=(xz, yz, z^2)

```python
def generateABC(xr, yr, zr):
    d={"X":[], "Y": [], "Z":[], "A": [],"B": [], "C": []}
    for i in range(1, xr+1):
        for j in range(i, yr+1):
            for k in range(j, zr+1):
                a=i*k
                b=j*k
                c=k**2
                if doesWork(a, b, c):
                    d["A"].append(a)
                    d["B"].append(b)
                    d["C"].append(c)
                    d["X"].append(i)
                    d["Y"].append(j)
                    d["Z"].append(k)
    return pd.DataFrame(d)
```

```python
DF2=generateABC(100, 100, 150)
display(DF2)
```

# 4.2 (c)
## write 4 primitive solutions
### primitive does not look like $ (n^{2}*A, n^{2}*B, n^{3}*C) $

```python
import math
def isPrimitive(a, b, c):
    for i in range(2, math.floor(math.sqrt(a))+1):
        if a%(i**2)==0 and b%(i**2)==0 and c%(i**3)==0:
            return False
    return True

def generatePrimitive(maxC):
    Idx=[]
    DF=generateTable(maxC, maxC, maxC)
    for index in DF.index:
        a=DF["A"][index]
        b=DF["B"][index]
        c=DF["C"][index]
        print(f'testing is primitive on {a}, {b}, {c}: {isPrimitive(a, b, c)}')
        if isPrimitive(a, b, c):
            Idx.append(index)
    d={"A": [], "B":[], "C":[]}
    for idx in Idx:
        d["A"].append(DF["A"][idx])
        d["B"].append(DF["B"][idx])
        d["C"].append(DF["C"][idx])
        
    return pd.DataFrame(d)

        
```

```python
DF=generatePrimitive(250)
display(DF)
```

## 4.2 (d)
### find primitive solutions where a=b

```python
def genSpecificPrim(maxC):
    d={"A":[], "B":[], "C":[]}
    for i in range(1, int(maxC/2)):
        for j in range(i, maxC):
            if doesWork(i, i, j):
                if isPrimitive(i, i, j):
                    d["A"].append(i)
                    d["B"].append(i)
                    d["C"].append(j)
            elif 2 * i**3 < j**2:
                break
    return pd.DataFrame(d)
```

```python
DF=genSpecificPrim(50000)
display(DF)
```

## 4.2 (e)
### find primitives where a>10000

```python
def generateBigPrimitives(minA): #ranges for each
    d={"A": [],"B": [], "C": []}
    for i in range(minA, minA*2):
        for j in range(i, minA*2):
            for k in range(j, int(math.sqrt(i**3+j**3))):
                if doesWork(i, j, k):
                    if isPrimitive(i, j, k):
                        print(f'working prim: {(i, j, k)}')
                        d["A"].append(i)
                        d["B"].append(j)
                        d["C"].append(k)
                elif i**3 + j**3 < k**2:
                    break
    return pd.DataFrame(d)
```

```python
DF=generateBigPrimitives(10000)
display(DF)
```
