```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def T(n):
    #returns the number of the Nth triangle
    return (n*(n+1))/2


p1=plt.plot(range(100),[T(x) for x in range(100)])
p2=plt.plot(range(100), [x**2 for x in range(100)])
```
